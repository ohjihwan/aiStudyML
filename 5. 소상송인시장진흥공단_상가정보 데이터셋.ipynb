git 은 100MB 의 파일은 레파지토리에 올라가지 못하도록 자동으로 막고 있습니다. 그런데 csv 파일이 너무 커서 올라 가지 않는 상황에서 업데이트를 위해 100MB 씩 쪼개서 올려야 하지만 파일의 손상의 위험이 있거나 데이터가 잘못 될 수 있는 가능성을 고려하여 방법을 찾던 도중 찾게 되었습니다.

### ※ Git LFS란?
**Git Large File Storage (Git LFS)** 의 줄임말로 큰 파일을 효율적으로 GitHub에서 관리하기 위한 확장 기능입니다.
Git은 원래 **소스 코드 관리용**이라 텍스트 중심으로 설계되어 있습니다.
그런데 이미지, 동영상, CSV, 모델 가중치처럼 100MB 이상이라면 바이너리 파일을 올릴 경우 저장소가 터지게 되어 있습니다.
이러한 문제를 해결하기 위해 나온게 Git LFS입니다.
LFS는 Git에 링크만 남기고, **실제 데이터는 외부 저장소(LFS 서버)에 둔다고 보면 됩니다.

### LFS 작동 원리
예를 들어 <mark> data.csv </mark>(300MB)를 커밋할 경우
1. <mark> git add data.csv </mark>
2. <mark> .gitattributes </mark> 파일에 <mark> *.csv filter=lfs </mark> 가 정의되어 있으면
3. Git은 진짜 파일 대신 **“포인터(pointer)”**를 저장합니다.

커밋된 포인터 파일 내용:
```
version https://git-lfs.github.com/spec/v1
oid sha256:abc123...
size 314572800
``` 

진짜 파일은 GitHub의 LFS 서버에 업로드됨.

clone 할 때는 Git이 자동으로
>“아, 이건 LFS 파일이네” <br>“그럼 LFS 서버에서 진짜 파일 받아야겠다”

하고 자동으로 받아옵니다.

### 설치 및 기본 사용법
**1단계: 설치**
```
git lfs install
```
**2단계: 추적할 파일 유형 지정**
```
git add data/big.csv
git commit -m "add large file"
git push origin main
```

이제 .csv는 LFS 서버에 올라갑니다.

### GitHub LFS 저장 한도

| 항목                 | 무료 계정 기준     |
| ------------------ | ------------ |
| 저장 공간              | 2GB          |
| 대역폭 (다운로드/업로드 트래픽) | 1GB/월        |
| 추가 구입              | 가능 ($5/50GB) |

LFS는 “용량 무제한”이 아닙니다.
GitHub는 LFS를 별도 공간으로 관리하므로, 한도가 넘으면 push가 실패하거나 느려집니다.

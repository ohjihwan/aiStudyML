```python
# EDA(탐색적 데이터 분석) 단계에서는 '정보 손실 방지'가 우선으로 인해 Cabin 값은 데이터 정보를 이용하되 유지 '원본 보존 vs 파생 변수' 전략
# Cabin 존재 여부는 정보가치가 있음
# 갑판 객실 위치에 앞글자만 처리, Cabin의 정보로 알 수 있는것
#     - A 갑판: 1등실 객실, 흡연실, 독서실, 산책 갑판 등이 있었습니다.
#     - B 갑판: 1등실 고급 객실, 고급 레스토랑, 산책 갑판 등이 있었습니다.
#     - C 갑판: 1등실 객실, 산책 갑판 등이 있었습니다.
#     - D 갑판: 1등실 객실, 2등실 객실, 선내 식당 등이 있었습니다.
#     - E 갑판: 2등실 객실, 3등실 객실, 그리고 방수문이 없는 복도가 있어 침몰 시 문제가 발생했던 구역이었습니다.
#     - F 갑판: 2등실 객실, 3등실 객실, 스쿼시 코트, 수영장, 터키탕 등이 있었습니다.
#     - G 갑판: 3등실 객실이 주로 위치했습니다.

# 1. Cabin 컬럼의 정보를 이용하여 앞글자만을 취급, Deck 추출
Titanic_df['Deck'] = Titanic_df['Cabin'].str[0]

# 2. NaN은 그대로 유지 (결측값)
# → 별도 처리 필요 없음. pandas에서 자동으로 NaN 처리됨

# 3. Has_Cabin 생성 (1: 있음, 0: 없음) 나중에 갑판 정보로 다른 결과를 얻을 수 있을지도 모름
Titanic_df['Has_Cabin'] = Titanic_df['Cabin'].notnull().astype(int)
```

```python
# 1. Deck 컬럼 생성 (앞에서 했음)
Titanic_df['Deck'] = Titanic_df['Cabin'].str[0]

# 2. Deck별 생존률 계산
deck_survival = Titanic_df.groupby('Deck')['Survived'].mean().sort_values(ascending=False)

# 3. Pandas 자체 시각화
deck_survival.plot(kind='bar', title='Deck - survival rate', figsize=(8, 5), color='skyblue')

# 4. 부가 설정
import matplotlib.pyplot as plt
plt.ylabel('survival Rate')
plt.xlabel('Deck')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Sex, Pclass, Embarked 등 범주형 변수의 고유값(unique), 빈도수(value_counts) 등을 분석

categorical_cols = ['Sex', 'Pclass', 'Embarked']

for col in categorical_cols:
    print(f"📌 [{col}] 변수 분석")
    print("고유값 (unique):", Titanic_df[col].unique())
    print("빈도수 (value_counts):")
    print(Titanic_df[col].value_counts(dropna=False))  # NaN 포함
    print("-" * 40)
```

```python
# 성별 생존률
sns.barplot(x='Sex', y='Survived', data=Titanic_df)
plt.title('Sex Survival Rate')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
# 여자가 더 높음! 구조 우선순위(“여성과 어린이 먼저”)가 실제 데이터에 반영됨
```

```python
# Pclass 생존률
sns.barplot(x='Pclass', y='Survived', data=Titanic_df)
plt.title('Pclass survival rate')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
# 높은 등급의 승객이 구조 시 더 유리한 위치에 있었을 가능성
```

```python
# Embarked 생존률
sns.barplot(x='Embarked', y='Survived', data=Titanic_df)
plt.title('Embarked survival Rate')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
# C > Q ≈ S 순 : C에서 탑승한 승객 중 1등석 비율이 높았을 가능성
# 항구 자체보다 탑승자의 사회적 지위 차이로 해석할 수 있음
```

```python
#  인코딩화, 주민번호 7자리를 인용하여 남자:1 / 여자:2 로 함
Titanic_df['Sex'] = Titanic_df['Sex'].map({'male': 1, 'female': 2})
#  인코딩화, index 기준
Titanic_df['Embarked'] = Titanic_df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
# 전처리된 데이터 저장
Titanic_df.to_csv("titanic_decked_data.csv", index=False)
```
